# 如何查看函数的属性

```python
from datetime import datetime
def log(level):
    def decorator(func):
        def wrapper(*args, **kw):
            print("{0} [{1}]:Call {2}|Param:({3},{4})".format(datetime.now(),level,func.__name__,args, kw))
            return func(*args, **kw)
        return wrapper
    return decorator

@log('ERROR')
def dective(*args, **kw):
    print("test function!")
dective(1,2,3)
```

```python
dir(log)
```
