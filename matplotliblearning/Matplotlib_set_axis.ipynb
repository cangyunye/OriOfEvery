# 设置轴
1. 设置轴取值范围xlim,ylim
2. 设置x,y,轴标签
3. 设置轴坐标范围定位点
4. 将轴坐标转换为文字

```python
import matplotlib.pyplot as plt
import numpy as np
def init_plot():
    x = np.linspace(-3,3,50)
    y1 = 2*x + 1
    y2  = x**2
    plt.figure(num=1)
    plt.plot(x,y2)
    plt.plot(x,y1,color='red',linewidth=1.0,linestyle='--')
init_plot()
plt.show()
```

## 设置轴取值范围xlim,ylim

```python
init_plot()
#设置取值范围
plt.xlim((-1,2))
plt.ylim((-2,3))
plt.show()
```

## 设置x,y,轴标签

```python
init_plot()
#轴标签名
plt.xlabel('Label X')
plt.ylabel('Label Y')
plt.show()
```

## 设置轴坐标范围定位点

```python
init_plot()
#轴坐标名
new_ticks = np.linspace(-1,2,5)
print(new_ticks)
plt.xticks(new_ticks)
plt.show()
```

## 将轴坐标转换为文字

```python
init_plot()
plt.yticks([-5,-2,-1,2,3,],['really bad','bad','normal','good','really good'])
plt.show()
```

```python
init_plot()
#改变字体，可以在字符串用r表示正则
plt.yticks([-5,-2,-1,2,3,],[r'$really\ bad$',r'$bad$',r'$normal$',r'$good$',r'$really\ good$'])
plt.show()
```

```python
init_plot()
#gca = 'get current axis'获取当前轴
ax = plt.gca()
#轴的脊梁，相对初始状态的4个边，以下将右边和上边置空
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#将底边作为X轴,左边作为Y轴
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
#将底边的(-1,1)作为坐标原点，即底边从(-1[normal])穿过，左边从1穿过
ax.spines['bottom'].set_position(('data',-1))
ax.spines['left'].set_position(('data',1))
plt.show()
```
