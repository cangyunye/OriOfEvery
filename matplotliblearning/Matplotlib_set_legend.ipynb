# 设置图例


```python
import matplotlib.pyplot as plt
import numpy as np
def init_plot():
    x = np.linspace(-3,3,50)
    y1 = 2*x + 1
    y2  = x**2
    plt.figure(num=1)
    #设置取值范围
    plt.xlim((-1,2))
    plt.ylim((-2,3))
    #轴标签名
    plt.xlabel('Label X')
    plt.ylabel('Label Y')
    return {'x':x,'y1':y1,'y2':y2}
```

```python
t=init_plot()

#设置图例标签
plt.plot(t['x'],t['y2'],label='up')
plt.plot(t['x'],t['y1'],color='red',linewidth=1.0,linestyle='--',label='down')
#绘制图例
plt.legend()

plt.show()
```

```python
t=init_plot()

#设置图例标签
l1,=plt.plot(t['x'],t['y2'],label='up')
l2,=plt.plot(t['x'],t['y1'],color='red',linewidth=1.0,linestyle='--',label='down')
#绘制图例,
plt.legend(handles=[l1,l2,],labels=['upup','downdown'],loc='best')#方位,可以upper right,使用best自动放在数据较少的地方

plt.show()
```
