# 方程图形化

```python
import matplotlib.pyplot as plt
import numpy as np
#x，-1到1，分50个点
x = np.linspace(-1,1,50)
y= 2*x + 1
print(x)
plt.plot(x,y)
plt.show()
```

```python
y= x**2 + 1
plt.plot(x,y)
plt.show()
```

# 多个图形figure，颜色，宽度，线型

```python
x = np.linspace(-1,1,50)
y1= 2*x + 1
y2= x**2
y3= x**2
#开启figure1
plt.figure()
#绘点
plt.plot(x,y1)
#开启figure2
plt.figure()#不给参数时，每次默认调用figure为自动升序
plt.plot(x,y2)
plt.figure(num=3,figsize=(8,5))#num为图形figure序号,figsize为长宽
#绘点使用红色，线宽2.0，"--"虚线
plt.plot(x,y3,color='red',linewidth=2.0,linestyle='--')
plt.show()
```
